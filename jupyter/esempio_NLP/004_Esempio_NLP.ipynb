```python
f = open("sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt", "r")
righe = f.readlines()
f.close()
```

```python
punteggiatura = [",", ".", "?", "!", "'", "-", "(", ")", ":", ";", "«", "»"]
```

```python
for i in range(len(righe)):
    righe[i] = righe[i].lower()
    for c in punteggiatura:
        righe[i] = righe[i].replace(c, " ")
        righe[i] = righe[i].strip()
```

```python
lista_parole = []
dizionario = {}
for riga in righe:
    riga.split()
    if(riga.split() != []):
        lista_parole = riga.split()
        for parola in lista_parole:
            if parola not in dizionario:
                dizionario[parola] = 1
            else:
                dizionario[parola] += 1
```
